```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['LotFrontage'] = (df['LotFrontage'].median())
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.dtypes
```

```python
if 'Id' in df.columns:
    df.drop('Id', axis=1, inplace=False)
```

```python
df['MSSubClass'] = df['MSSubClass'].astype(str)
df['MoSold'] = df['MoSold'].astype(str)
df['YrSold'] = df['YrSold'].astype(str)
```

```python
print(df[['MSSubClass', 'MoSold', 'YrSold']].dtypes)
```

```python
df.shape
```

```python
df = pd.get_dummies(df)
```

```python
df.shape
```

```python
y= df['SalePrice']
x = df.drop('SalePrice', axis=1)
```

```python
print("x shape: ", x.shape)

print("y shape: ", y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=42)

print("x_train", x_train)
print("y_train", y_train)
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
prediction = model.predict(x_test)
```

```python
mean_absolute_error(y_test,prediction)
```

```python
acuracy_score = r2_score(y_test,prediction)
print(acuracy_score)
```

```python


test_df = pd.read_csv('test.csv')

# ID bachana
test_ids = test_df['Id']
test_df.drop('Id', axis=1, inplace=True)

test_df['MSSubClass'] = test_df['MSSubClass'].astype(str)
test_df['MoSold'] = test_df['MoSold'].astype(str)
test_df['YrSold'] = test_df['YrSold'].astype(str)

# --- NAYA UPDATED FOR LOOP --- 
# Ab hum safely check karenge ki Number hai ya nahi
for col in test_df.columns:
    if pd.api.types.is_numeric_dtype(test_df[col]):
        # Agar strictly number hai, tabhi Median lagana
        test_df[col].fillna(test_df[col].median(), inplace=True)
    else:
        # Agar number chhod kar kuch bhi hai (object, string, text), toh Mode lagana
        if not test_df[col].mode().empty:
            test_df[col].fillna(test_df[col].mode()[0], inplace=True)
        else:
            test_df[col].fillna("Missing", inplace=True)
# ------------------------------

# 4. Encoding
test_encoded = pd.get_dummies(test_df)
# 5. Columns Align karna
missing_cols = set(x.columns) - set(test_encoded.columns)
for c in missing_cols:
    test_encoded[c] = 0

test_encoded = test_encoded[x.columns]

# 6. Predict & Save
test_predictions = model.predict(test_encoded)
submission = pd.DataFrame({'Id': test_ids, 'SalePrice': test_predictions})
submission.to_csv('my_kaggle_submission.csv', index=False)

pd.read_csv('my_kaggle_submission.csv').head()


```
